# Clustering with hierarchical trees

## Grouping clusters in bottom-up fashion

```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


url="company-stock-movements-2010-2015-incl.csv"
df = pd.read_csv(url)
labels = df["Unnamed: 0"].tolist()
df = df.drop(["Unnamed: 0"],axis=1)
df.head()
```

```python

X = []
for columna in df:
    X.append(df[columna].tolist())
len(X)
```

```python
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt
import numpy as np
fn = 'company-stock-movements-2010-2015-incl.csv'
stocks_df = pd.read_csv(fn, index_col=0)
companies = list(stocks_df.index)
movements = stocks_df.values
```

## Buscar el millor nombre de clusters

```python
X_norm=Normalizer().fit_transform(movements)

distortions = []
for i in range(5, 25):
    km = KMeans(n_clusters=i, 
                init='k-means++', 
                n_init=100, 
                max_iter=300, 
                random_state=0)
    km.fit(X_norm)
    distortions.append(km.inertia_)
plt.plot(range(5, 25), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.savefig('Elbow.png', dpi=300)
plt.show()
```

```python
nc=10
km = KMeans(n_clusters=nc,init='k-means++', n_init=100, max_iter=3000, random_state=0)
km.fit(X_norm)
```

```python
labels2 = km.predict(movements)

df2 = pd.DataFrame({'labels': labels2, 'companies': companies})
df2.sort_values('labels')
```

```python
from scipy.spatial.distance import pdist, squareform
import seaborn as sns

distances=squareform(pdist(X_norm))
y_km = labels2
inds=np.argsort(y_km)
dist = pd.DataFrame(distances[inds])
sns.heatmap(dist[inds],yticklabels=y_km[inds])
plt.show()
```

## Performing hierarchical clustering unsing scipy on a distance matrix

https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html


```python
from scipy.spatial.distance import pdist, squareform

row_dist = pd.DataFrame(squareform(pdist(X_norm, metric='euclidean')),
                        columns=labels,
                        index=labels)
row_dist
```

We can either pass a condensed distance matrix (upper triangular) from the `pdist` function, or we can pass the "original" data array and define the `metric='euclidean'` argument in `linkage`. However, we should not pass the squareform distance matrix, which would yield different distance values although the overall clustering could be the same.

```python
# 2. correct approach: Condensed distance matrix
from scipy.cluster.hierarchy import linkage

row_clusters = linkage(pdist(X_norm, metric='euclidean'), method='complete')
pd.DataFrame(row_clusters,
             columns=['row label 1', 'row label 2',
                      'distance', 'no. of items in clust.'],
             index=['cluster %d' % (i + 1) 
                    for i in range(row_clusters.shape[0])])
```

```python
from scipy.cluster.hierarchy import dendrogram


# make dendrogram black (part 1/2)
# from scipy.cluster.hierarchy import set_link_color_palette
# set_link_color_palette(['black'])
row_dendr = dendrogram(row_clusters,
                       labels=labels,
                       # make dendrogram b)lack (part 2/2)
                       # color_threshold=np.inf
                       )
plt.tight_layout()
plt.ylabel('Euclidean distance')
#plt.savefig('11_11.png', dpi=300,
#            bbox_inches='tight')
plt.show()
```

## Applying agglomerative clustering via scikit-learn

http://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html


```python
from sklearn.cluster import AgglomerativeClustering
nclus = 10

ac = AgglomerativeClustering(n_clusters=nclus, 
                             affinity='euclidean', 
                             linkage='complete')
labels2 = ac.fit_predict(X_norm)
print('Cluster labels: %s' % labels2)
print(len(df))
df2 = pd.DataFrame({'labels': labels2, 'companies': companies})
df2.sort_values('labels')
```

## Finding parameters in DBScan

```python
import numpy as np

from sklearn.metrics.pairwise import pairwise_distances
from scipy.spatial.distance import pdist, squareform

distances=squareform(pdist(X))

kcloser = 2
lkthc = [np.sort(x)[kcloser+1] for x in distances]
```

```python
plt.plot(np.sort(lkthc))
plt.show()
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=2, min_samples=kcloser, metric='euclidean')
y_db = db.fit_predict(X)

lc = ['lightblue','red','yellow','green','black','brown','blue','orange','purple','brown']

for i in range(max(y_db)+1):
    plt.scatter(X[y_db == i, 0], X[y_db == i, 1],
                c=lc[i], marker='o', s=40,
                edgecolor='black', 
                label='cluster '+str(i))
    
plt.scatter(X[y_db < 0, 0], X[y_db <0, 1],
                c='white', marker='o', s=40,
                edgecolor='black', 
                label='noise')
plt.title("DBSCAN")
plt.tight_layout()
#plt.savefig('images/11_16.png', dpi=300)
plt.show()
y_db
```
